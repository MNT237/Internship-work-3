```python






















from sklearn.cluster import KMeans
import pandas as pd

# The same 6 students from the board
df = pd.DataFrame({
    'student' : ['A','B','C','D','E','F'],
    'math'    : [80, 75, 85, 30, 25, 35],
    'science' : [85, 80, 90, 25, 30, 20]
})

X = df[['math','science']]

# K=2 — find 2 groups
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)

df['group'] = kmeans.labels_
print(df)

print(f"\nCenter 1: {kmeans.cluster_centers_[0]}")
print(f"Center 2: {kmeans.cluster_centers_[1]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='math', y='science', hue='group', data=df, palette='viridis', s=100, alpha=0.8)

# Plot the cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            marker='X', s=200, color='red', label='Cluster Centers')

plt.title('K-Means Clustering of Student Scores (K=2)')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 1. Setup PyCaret Environment

First, we need to set up the PyCaret environment. This function prepares the data and deployment environment. It performs all the data preprocessing steps and then creates a transformation pipeline to prepare the data for modeling. We will use the `df` DataFrame created earlier.

```python
from pycaret.clustering import setup, create_model, assign_model, plot_model

# Setup the environment
# We exclude 'student' column as it's an identifier, not a feature for clustering
# session_id for reproducibility
# We also don't scale the data by default for this tutorial, but it's often a good practice.
clustering_setup = setup(data=df.drop('student', axis=1), session_id=42, verbose=False)
```

### 2. Create a Clustering Model

Now, let's create a K-Means clustering model with 2 clusters, similar to our manual example. PyCaret allows you to easily create various clustering models with a single function call.

```python
# Create a K-Means model with 2 clusters
kmeans_pycaret = create_model('kmeans', num_clusters=2)

print(kmeans_pycaret)
```

### 3. Assign Clusters to Original Data

After creating the model, we can assign the predicted clusters back to our original DataFrame.

```python
# Assign clusters to the original data
df_clustered = assign_model(kmeans_pycaret)
display(df_clustered.head())
```

### 4. Visualize the Clusters

PyCaret also provides built-in plotting functions to visualize the clusters directly. We can use the 'cluster' plot to see how the data points are grouped.

```python
# Plot the clusters (2D Projection using t-SNE, PCA or UMAP if features > 2)
# Since we have only two features (math, science), it will directly plot them.
plot_model(kmeans_pycaret, plot='cluster')
```

### Exploring HDBSCAN Clustering

HDBSCAN (Hierarchical Density-Based Spatial Clustering of Applications with Noise) is a powerful clustering algorithm that extends DBSCAN. It's particularly good at finding clusters of varying densities and identifying noise points. A key advantage is that it doesn't require you to specify the number of clusters beforehand.

First, we need to install the `hdbscan` library.

```python
!pip install hdbscan -q
```

```python
# The HDBSCAN model fitting and group assignment code has been moved to the plotting cell for robustness.
```

HDBSCAN assigns `-1` to noise points (outliers). In our small dataset, it should still form two clear groups. Now, let's visualize these clusters.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import hdbscan

# Apply HDBSCAN to the data
# min_cluster_size: The minimum number of samples in a cluster.
# We'll use 2 as we expect small clusters.
hdbscan_model = hdbscan.HDBSCAN(min_cluster_size=2, min_samples=1, cluster_selection_epsilon=0.5)
hdbscan_model.fit(X)

# Add the HDBSCAN cluster labels to the DataFrame
df['hdbscan_group'] = hdbscan_model.labels_

plt.figure(figsize=(8, 6))
sns.scatterplot(x='math', y='science', hue='hdbscan_group', data=df, palette='viridis', s=100, alpha=0.8)

plt.title('HDBSCAN Clustering of Student Scores')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.legend(title='HDBSCAN Group')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
